```python
x = np.linspace(0,6*np.pi,200)
```

```python
s = np.sin(x)
c = np.cos(x)
```

```python
plt.plot(x,s)
```

```python
plt.plot(x,s,color="r",label="sin(x)")
plt.plot(x,c,color="g",label="cos(x)")
plt.ylabel("y")
plt.xlabel("x")
plt.legend()
```
