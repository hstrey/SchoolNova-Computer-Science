```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,11)
x
```

```python
y = 2*x + np.random.normal(loc=0, scale =1,size=11)
```

```python
plt.plot(x,y,"o")
```

```python
p = np.polyfit(x,y,1)
print(p)
```

```python
yfit = x*p[0] + p[1]
plt.plot(x,y,"o")
plt.plot(x,yfit)
```

```python
y = 2*x + 0.1*x**2 + np.random.normal(loc=0, scale =1,size=11)
```

```python
plt.plot(x,y,"o")
```

```python
p = np.polyfit(x,y,1)
p2 = np.polyfit(x,y,2)
p10 = np.polyfit(x,y,10)
```

```python
xf = np.linspace(-5,5,200)
yfit = p[0]*xf + p[1]
yfit2 = p2[0]*xf**2 + p2[1]*xf + p2[2]
yfit10 = np.poly1d(p10)(xf)
```

```python
plt.plot(x,y,"o")
plt.plot(xf,yfit)
plt.plot(xf,yfit2)
plt.plot(xf,yfit10)
```
