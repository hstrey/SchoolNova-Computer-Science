```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
x = np.random.random(N)*2
y = np.random.random(N)*2
r = np.sqrt((x-1)**2+(y-1)**2)
Nc = 0
Nc_list_x = []
Nc_list_y = []
Nnc_list_x =[]
Nnc_list_y=[]
for i,radius in enumerate(r):
    if radius<=1:
        Nc = Nc +1
        Nc_list_x.append(x[i])
        Nc_list_y.append(y[i])
    else:
        Nnc_list_x.append(x[i])
        Nnc_list_y.append(y[i])
```

```python
print("Our estimate of pi: ",Nc*4/N)
```

```python
plt.scatter(Nc_list_x,Nc_list_y)
plt.scatter(Nnc_list_x,Nnc_list_y)
```
