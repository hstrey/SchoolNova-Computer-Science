```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# this is an example to explain what zip does
l = [4,5,6,7,8]
m = [9,1,2,3,4]
list(zip(l,m))
```

```python
N = 1000
x = np.random.random(N)*2
y = np.random.random(N)*2
r = np.sqrt((x-1)**2+(y-1)**2)
Nc = 0
Nc_list_x = []
Nc_list_y = []
Nnc_list_x =[]
Nnc_list_y=[]
for xx,yy,radius in zip(x,y,r):
    if radius<=1:
        Nc = Nc +1
        Nc_list_x.append(xx)
        Nc_list_y.append(yy)
    else:
        Nnc_list_x.append(xx)
        Nnc_list_y.append(yy)
```

```python
print("Our estimate of pi: ",Nc*4/N)
```

```python
print(np.pi)
```

```python
plt.scatter(Nc_list_x,Nc_list_y)
plt.scatter(Nnc_list_x,Nnc_list_y)
```

```python
N = 1000
x = np.random.random(N)*2
y = np.random.random(N)*2
r = np.sqrt((x-1)**2+(y-1)**2)
xc = x[r<=1]
yc = y[r<=1]
xnc = x[r>1]
ync = y[r>1]
len(xc)
```

```python
print("estimated pi ",4*len(xc)/N)
```

```python
plt.scatter(xc,yc)
plt.scatter(xnc,ync,color="r")
plt.axis('square')
plt.xlim((0,2))
plt.ylim((0,2))
circle1 = plt.Circle((1,1),1,fill=False)
ax = plt.gca()
ax.add_artist(circle1)
plt.savefig('circle.pdf')
```
